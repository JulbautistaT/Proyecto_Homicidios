```python
import pandas as pd
import plotly.express as px
```

```python
# Cargar el archivo generado anteriormente
df_final = pd.read_csv('/content/Homicidios_limpio.csv')
df_final.head()
```

```python
# Asegurarse de que las fechas están en formato datetime si es necesario
df_final['Fecha_hecho'] = pd.to_datetime(df_final['Fecha_hecho'])
```

```python

# Gráfico 1: Homicidios por Departamento
def plot_homicidios_por_departamento(df):
    homicidios_por_departamento = df['Departamento'].value_counts().reset_index()
    homicidios_por_departamento.columns = ['Departamento', 'Cantidad']

    fig = px.bar(homicidios_por_departamento,
                 x='Departamento', y='Cantidad',
                 title="Gráfico 1: Homicidios por Departamento",
                 labels={'Departamento': 'Departamento', 'Cantidad': 'Cantidad de Homicidios'},
                 color='Cantidad',
                 color_continuous_scale='Blues')

    fig.update_layout(xaxis_title='Departamento', yaxis_title='Cantidad de Homicidios',
                      xaxis_tickangle=-45)

    fig.show()
```

```python
# Gráfico 2: Distribución de Homicidios por Género
def plot_homicidios_por_genero(df):
    homicidios_por_genero = df['Genero'].value_counts().reset_index()
    homicidios_por_genero.columns = ['Género', 'Cantidad']

    fig = px.pie(homicidios_por_genero, values='Cantidad', names='Género',
                 title="Gráfico 2: Distribución de Homicidios por Género",
                 color_discrete_sequence=px.colors.sequential.RdBu)

    fig.show()
```

```python
# Gráfico 3: Tendencia de Homicidios por Año
def plot_tendencia_homicidios(df):
    df['Año'] = df['Fecha_hecho'].dt.year
    homicidios_por_año = df.groupby('Año').size().reset_index(name='Cantidad')

    fig = px.line(homicidios_por_año, x='Año', y='Cantidad',
                  title="Gráfico 3: Tendencia de Homicidios por Año",
                  labels={'Año': 'Año', 'Cantidad': 'Cantidad de Homicidios'},
                  markers=True)

    fig.update_traces(line=dict(color='royalblue', width=4))
    fig.show()
```

```python
# Gráfico 4: Homicidios por Grupo Etario
def plot_homicidios_por_grupo_etario(df):
    homicidios_por_grupo = df['Grupo_etario'].value_counts().reset_index()
    homicidios_por_grupo.columns = ['Grupo Etario', 'Cantidad']

    fig = px.bar(homicidios_por_grupo,
                 x='Grupo Etario', y='Cantidad',
                 title="Gráfico 4: Homicidios por Grupo Etario",
                 labels={'Grupo Etario': 'Grupo Etario', 'Cantidad': 'Cantidad de Homicidios'},
                 color='Cantidad',
                 color_continuous_scale='Greens')

    fig.update_layout(xaxis_title='Grupo Etario', yaxis_title='Cantidad de Homicidios')
    fig.show()
```

```python
# Gráfico 5: Homicidios por Medios (ARMAS MEDIOS)
def plot_homicidios_por_medios(df):
    homicidios_por_medios = df['Armas_medios'].value_counts().reset_index()
    homicidios_por_medios.columns = ['Armas/Medios', 'Cantidad']

    fig = px.bar(homicidios_por_medios,
                 x='Armas/Medios', y='Cantidad',
                 title="Gráfico 5: Homicidios por Medios",
                 labels={'Armas/Medios': 'Armas/Medios', 'Cantidad': 'Cantidad de Homicidios'},
                 color='Cantidad',
                 color_continuous_scale='Oranges')

    fig.update_layout(xaxis_title='Armas/Medios', yaxis_title='Cantidad de Homicidios',
                      xaxis_tickangle=-45)
    fig.show()
```

```python
# Gráfico 6: Homicidios por Mes del Año
def plot_homicidios_por_mes(df):
    df['Mes'] = df['Fecha_hecho'].dt.month
    homicidios_por_mes = df.groupby('Mes').size().reset_index(name='Cantidad')

    fig = px.line(homicidios_por_mes, x='Mes', y='Cantidad',
                  title="Gráfico 6: Homicidios por Mes del Año",
                  labels={'Mes': 'Mes', 'Cantidad': 'Cantidad de Homicidios'},
                  markers=True)

    fig.update_traces(line=dict(color='darkorange', width=4))
    fig.show()
```

```python
# Gráfico 7: Los 5 Municipios con Mayor Cantidad de Homicidios
def plot_top_5_municipios(df):
    homicidios_por_municipio = df['Municipio'].value_counts().reset_index()
    homicidios_por_municipio.columns = ['Municipio', 'Cantidad']
    top_5_municipios = homicidios_por_municipio.head(5)

    fig = px.bar(top_5_municipios,
                 x='Municipio', y='Cantidad',
                 title="Gráfico 7: Los 5 Municipios con Mayor Cantidad de Homicidios",
                 labels={'Municipio': 'Municipio', 'Cantidad': 'Cantidad de Homicidios'},
                 color='Cantidad',
                 color_continuous_scale='Reds')

    fig.update_layout(xaxis_title='Municipio', yaxis_title='Cantidad de Homicidios',
                      xaxis_tickangle=-45)
    fig.show()
```

```python
# Ejecutar los gráficos
plot_homicidios_por_departamento(df_final)
plot_homicidios_por_genero(df_final)
plot_tendencia_homicidios(df_final)
plot_homicidios_por_grupo_etario(df_final)
plot_homicidios_por_medios(df_final)
plot_homicidios_por_mes(df_final)
plot_top_5_municipios(df_final)
```
