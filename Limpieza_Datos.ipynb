## Importar las librerías necesarias

## Normalizando las columnas y filas del dataset

```python
# Normalizar nombres de columnas
df.columns = df.columns.str.capitalize()
df.columns
```

```python
#Cambiar columnas que tienen espacios y tildes
df=df.rename(columns={
    'Codigo dane': 'Codigo_dane',
    'Fecha hecho': 'Fecha_hecho',
    'Grupo etarío': 'Grupo_etario',
    'Descripción conducta':'Descripcion_conducta',
    'Armas medios':'Armas_medios'

})
```

```python
df
```

```python
#df[df.select_dtypes(['object']).columns]

# Aplicar la capitalización a cada una de las columnas con el tipo object de texto
for col in df.select_dtypes(['object']).columns:
    df[col] = df[col].str.capitalize()
```

```python
df_sorted = df.sort_values(by='Departamento')
```

```python
df
```

## **Tratando duplicados y valores nulos**

```python
#Identififcar si hay valores nulos
df.isnull().sum()
```

```python
#Cuantos duplicados hay
df.duplicated().sum()
```

```python
# Identificar cuales son los duplicados
duplicados = df[df.duplicated()]
duplicados
```

## **Eliminación de columnas sin uso o con información poco relevante para el caso de estudio**




```python
for col in df.columns:
  print(col,np.sum(pd.isnull(df[col])))
```

```python
# Eliminar columna Descripcion_conducta
df = df.drop(columns=['Descripcion_conducta', 'Codigo_dane'])
```

```python
# Visualizar el dataframe nuevamente
df.head()
```

```python
# Para interes del proyecto queremos que la base de datos sea de la última decada

df['Fecha_hecho'] = pd.to_datetime(df['Fecha_hecho'], errors='coerce')

# Obtener el año actual
anio_actual = pd.Timestamp.now().year

# Filtrar los datos de los últimos 10 años
df = df[df['Fecha_hecho'].dt.year >= (anio_actual - 10)]

# Formatear la columna 'Fecha_hecho' al formato '%d/%m/%Y'
df['Fecha_hecho'] = df['Fecha_hecho'].dt.strftime('%d/%m/%Y')
```

```python
df.head()
```

## **Verificar y guardar Dataset limpio**

```python
df_limpio=df.drop_duplicates()
```
